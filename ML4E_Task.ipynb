```python
import pandas as pd
trainset = pd.read_csv('train.csv')
trainset = trainset.dropna()
trainset.describe()
```

```python
trainset.head()
```

```python
x_train = trainset[trainset.columns[1:9]]
x_train.head()
```

```python
y_train = trainset[trainset.columns[-1]]
y_train
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
model = RandomForestClassifier()
param_grid = {'criterion' :('gini','entropy')}
model_cv = GridSearchCV(model, param_grid,scoring='accuracy', cv=5)
model_cv.fit(x_train, y_train)
```

```python
testset = pd.read_csv('test.csv')
testset = testset.dropna()
testset.head()
```

```python
x_test = testset[testset.columns[1:9]]
y_pred = model_cv.predict(x_test)
```

```python
output = {'ID' : testset['ID'], 'Crop_Damage' : y_pred}
output = pd.DataFrame(output)
output
```

```python
output.to_csv('sample_submission.csv', index=False)
```
